# Read Data

```python
import pandas as pd

df = pd.read_csv('data.csv')

print(df.head(5))

ACCURACY = {'SVM':0.0, 'RF':0.0}
```

# Data Processing

Checking Null values

```python
df.isnull().values.any()
```

Finding impacts of the input columns

```python
import seaborn as sns
%matplotlib inline

# Import label encoder
from sklearn import preprocessing
import matplotlib.pyplot as plt
 
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'ID'.
df['ID']= label_encoder.fit_transform(df['ID'])
  
df['ID'].unique()

df.corr()
```

Create new Dataframe consisting most impactful column having over 60% correlation with Grade

```python
new_df = df[['Week2_Quiz1', 'Week3_MP1','Week3_PR1', 'Week5_MP2', 'Week5_PR2', 'Week7_MP3' , 'Week7_PR3', 'Week4_Quiz2', 'Week6_Quiz3' , 'Week3_Stat0', 'Week4_Stat0', 'Week4_Stat1', 'Week6_Stat0', 'Week6_Stat1', 'Grade']] 
print(new_df)
```

Heatmap of update Dataframe

```python
fig, ax = plt.subplots(figsize=(50,50))

sns.heatmap(new_df.corr(), annot=True, square=True,annot_kws={"fontsize":18})
```

# Training and Test Dataset

```python
X = new_df.drop('Grade', axis = 1)
y = new_df['Grade']

# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,random_state=109) # 80% training and 20% test
```

# Support Vector Machine

---



# Train the Model


```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

# Performance Evaluation

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
accuracy = metrics.accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

ACCURACY["SVM"] = (round(accuracy, 2))

from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Random Forest

---





# Train The Model

```python
# importing random forest classifier from assemble module
from sklearn.ensemble import RandomForestClassifier

# creating a RF classifier
clf_RF = RandomForestClassifier(n_estimators = 100, random_state= 30) 
 
# Training the model on the training dataset
# fit function is used to train the model using the training sets as parameters
clf_RF.fit(X_train, y_train)
 
# performing predictions on the test dataset
y_pred = clf_RF.predict(X_test)
 
```

# Performance evaluation

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
ACCURACY["RF"] = (round(metrics.accuracy_score(y_test, y_pred),2))

from sklearn.metrics import confusion_matrix, classification_report
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

# Plot Bar Chart of Accuracy

```python
print(ACCURACY)
models = ACCURACY.keys()
accuracies = ACCURACY.values()
  
fig = plt.figure(figsize = (5, 5))
 
# creating the bar plot
plt.bar(models, accuracies)
 
plt.xlabel("Machine Learning Models")
plt.ylabel("Accuracy")
plt.title("Acuuracy of Machine Learning Models")
plt.show()
```

# Feature Importance

---



```python
from matplotlib import pyplot
import numpy as np

# get importance
importance = clf_RF.feature_importances_
x_label = list(new_df.columns.values.tolist())

print (importance)

#Sort Feature importances in descending order
indices = np.argsort(importance)[::-1]

#Rearrange feature names so they match the feature importances
names = [x_label[i] for i in indices] 

#Create plot
plt.figure()
f, ax = plt.subplots(figsize=(18,10))
plt.title('Feature Importance')
plt.bar(range(X.shape[1]), importance[indices])
plt.xticks(range(X.shape[1]), names, rotation=90)
plt.show()
```
